```python
df = pd.read_csv('10000 Rows of Parking Violations.txt')
more_columns_to_drop = [
    'Summons Number',
    'Plate ID',
    'Plate Type',
    'Vehicle Make',
    'Street Code1',
    'Street Code2',
    'Street Code3',
    'Violation Precinct',
    'Issuer Precinct',
    'Issuer Code',
    'Issuer Command',
    'Issuer Squad',
    'Time First Observed',
    'Violation In Front Of Or Opposite',
    'House Number',
    'Street Name',
    'Intersecting Street',
    'Law Section',
    'Sub Division',
    'From Hours In Effect',
    'To Hours In Effect',
    'Double Parking Violation',
    'Hydrant Violation',
    'No Standing or Stopping Violation',
    'Violation Description',

]

df = df.drop(columns=more_columns_to_drop)
df['Violation Code'] = df['Violation Code'].astype(str)
df['Violation Code'].describe()


#The most common violation was code 98: Standing or parking in front of a public or private driveway. The owner or renter of a lot accessed by a private driveway may park a passenger vehicle registered to him / her at that address in front of the driveway provided the lot does not contain more than 2 dwelling units and that parking does not violate any other rule or restriction.
#https://www.nyc.gov/site/finance/vehicles/services-violation-codes.page


violation_counts = df['Violation Code'].value_counts()
violation_counts = violation_counts.sort_index()
sns.barplot(x=violation_counts.index, y=violation_counts.values)
plt.xlabel('Violation Code')
plt.ylabel('Count')
plt.title('Count of Violations by Code')
plt.xticks(rotation=90)
plt.show()
```
